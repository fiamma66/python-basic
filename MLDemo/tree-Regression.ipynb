```python

# 載入套件 & 資料集
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
import pandas as pd
import graphviz
```

```python
# load_iris() 為一字典 資料
boston = load_boston()
df = pd.DataFrame(boston["data"],columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```python
# 切開 --> 做準備
# 用sklearn內建 split
from sklearn.model_selection import train_test_split

# 傳回四組 array
# [90%題目, 10%題目,  90%答案 , 10%答案]
x_train, x_test, y_train, y_test=train_test_split(boston["data"], boston["target"], test_size=0.1)


```

```python
# 算出方程式 --> fit 訓練模型 90% 資料
# max_depth = 2 --> 最多往下長 2 層
reg = DecisionTreeRegressor(max_depth=4)
# 帶入90%題目 90%答案
reg.fit(x_train, y_train)
```

```python
# 將決策樹 畫出來
from sklearn.tree import export_graphviz
g = export_graphviz(reg, feature_names=boston["feature_names"], 
                         filled=True)

graphviz.Source(g)
```

```python

# 驗證方程式的效果 --> 測試模型 10% 資料
pre = reg.predict(x_test)
print("預測 : ", list(pre))
print("正確 : ", list(y_test))


# 每次都會不一樣 ? WHY?
# 每次切割 都是隨機 結果會不同
```

```python
from sklearn.metrics import r2_score
# mse 來代表雜亂程度 
# 以平均來區分
# 回歸使用 r2_score(y_true,predict)

print("r2 score:", r2_score(y_test, pre))
```
