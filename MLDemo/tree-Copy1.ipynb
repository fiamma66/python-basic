```python

# 載入套件 & 資料集
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd
import graphviz
```

```python
# load_iris() 為一字典 資料
iris = load_iris()
df = pd.DataFrame(iris["data"],columns=iris["feature_names"])
df["target"] = iris["target"]

df
```

```python
# 切開 --> 做準備
# 用sklearn內建 split
from sklearn.model_selection import train_test_split

# 傳回四組 array
# [90%題目, 10%題目,  90%答案 , 10%答案]
x_train, x_test, y_train, y_test=train_test_split(iris["data"], iris["target"], test_size=0.1)


```

```python
# 事前知道 iris 有 3 種
clu = KMeans(n_clusters=3)
# 實際分群時 根本不知道 y 
# 所以只帶入 x
clu.fit(x_train)
# init='k-means++' 
# 上表示 一開始不會選太接近的圓心
```

```python
print("預測 : ",list(clu.predict(x_test)))
```

```python
print("本來的標籤 : ",list(y_test))
```

```python
# 預測的正確率計算
# 使用混淆矩陣
from sklearn.metrics import accuracy_score

accuracy_score(clu.predict(x_test), y_test)
```

```python
# 拿出資料點被打上的標籤
clu.labels_
```

```python
# n_clusters = 1 是無效的 只分一群到底是 ?
# 從 2 開始
from sklearn.metrics import silhouette_score

xs = []
ys = []

for k in range(2,6):
    clu = KMeans(n_clusters=k)
    clu.fit(iris["data"])
    xs.append(k)
    # silhourtte_score(X,labels)
    score = silhouette_score(iris["data"], 
                             clu.labels_)
    ys.append(score)
    
print(xs)
print(ys)

# 2群最高準確 why?
# iris 中 有兩個非常相似 另外一種有明顯區別
```

```python
# 簡易作圖 matplotlib
# 只需傳入 (x , y)
import matplotlib.pyplot as plt

# 只要加入這行 可以自動秀圖
# only for jupyter-notebook
%matplotlib inline
# 盡量不要使用 中文 做標籤 或 標題
plt.title("Sihouette Score")
# plot(x,y, format="[color][marker][line]")
plt.xlabel("K")
plt.ylabel("Score")
plt.plot(xs, ys, "m^--")



# jupyter notebook 可以省略 plt.show()
#plt.show()
# 使用 pycharm 則必須使用 plt.show()


```
